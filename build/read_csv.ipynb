```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML
```

```python
import itertools
import matplotlib.pyplot as plt
import matplotlib.animation as animation

nbr_outputs = 50

fig, ax = plt.subplots()
i  = 0
df = pd.read_csv("CSV/data_{0}.csv".format(i+1))
x, y = df["pos_x"],df["pos_y"]
mat, = ax.plot(x, y, 'o')


def animate(i):
    df = pd.read_csv("CSV/data_{0}.csv".format(i+1))
    x, y = df["pos_x"],df["pos_y"]
    mat.set_data(x, y)
    return mat,

ax.axis()
ani = animation.FuncAnimation(fig, animate, frames=nbr_outputs,interval=200)
plt.show()
HTML(ani.to_jshtml())
```
