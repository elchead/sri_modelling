```python
import pandas as pd
from IPython.display import HTML
import numpy as np
```

```python
import itertools
import matplotlib.pyplot as plt
import matplotlib.animation as animation

nbr_outputs = 100

fig, ax = plt.subplots()
i  = 0
df = pd.read_csv("./build/CSV/data_{0}.csv".format(i+1))
x, y = df["pos_x"],df["pos_y"]
mat = ax.scatter(x, y)


def animate(i):
    df = pd.read_csv("./build/CSV/data_{0}.csv".format(i+1))
    x, y = df["pos_x"],df["pos_y"]
    state = df["state"]
    c = ['b' if s==0 else 'r' if s==1 else 'g' for s in state]
    label = ['susceptible' if s==0 else 'infected' if s==1 else 'recovered' for s in state]
    mat.set_color(c)
    mat.set_offsets(np.c_[x,y])
    return mat

ax.axis()
ax.set_xlim([-1.1,1.1])
ax.set_ylim([-1.1,1.1])
ani = animation.FuncAnimation(fig, animate, frames=nbr_outputs,interval=180)
plt.show()
HTML(ani.to_jshtml())
```
